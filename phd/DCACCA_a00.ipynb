Space-time diagrams of an one-dimensional density-conserving rule parameterized by $(0.5,0,0)$ and a configuration with  and without substrings of the form $(0)^j$, $j>1$

```python
import catools as catools
import matplotlib.pyplot as plt
```

```python
a = 0.5
lut = catools.density_conserving_1d_lut(a, 0, 0)
catools.display_lut(lut)
```

```python
configuration_with_substrings00 = catools.parse_1d_configuration("000101010111110000000000001100000")
configuration_with_no_substrings00 = catools.parse_1d_configuration("101101010111110101010101011101010")

spacetime_of_configuration_with_substrings00 = list(catools.iterate_1dconfiguration_to(configuration_with_substrings00, lut, 50))
spacetime_of_configuration_with_no_substrings00 = list(catools.iterate_1dconfiguration_to(configuration_with_no_substrings00, lut, 50))

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(10, 10)
catools.plot_spacetime(spacetime_of_configuration_with_substrings00, ax=ax1)
catools.plot_spacetime(spacetime_of_configuration_with_no_substrings00, ax=ax2)
```

Below, space time diagram of a configuration without substrings presented above and the same configuration with a small modification that produces a configuration with a substring $00$.

```python
configuration_with_substrings00 = catools.parse_1d_configuration("101101010111100101010101011101010")
configuration_with_no_substrings00 = catools.parse_1d_configuration("101101010111110101010101011101010")

spacetime_of_configuration_with_substrings00 = list(catools.iterate_1dconfiguration_to(configuration_with_substrings00, lut, 50))
spacetime_of_configuration_with_no_substrings00 = list(catools.iterate_1dconfiguration_to(configuration_with_no_substrings00, lut, 50))

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(10, 10)
catools.plot_spacetime(spacetime_of_configuration_with_substrings00, ax=ax1)
catools.plot_spacetime(spacetime_of_configuration_with_no_substrings00, ax=ax2)
```
