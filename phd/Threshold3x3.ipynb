```python
import catools as catools
```

```python
lut = catools.num([0.0094220788056185,0.0089778109738877,0.0506542601172838,0.0148475414685278,0.3582985356212308,0,0.4368476624947731,0.0349355409111639,0.1542649789427312,0.0564312532604916,0.7700410565079709,0.3954581037799654,0.4888848643107001,0.9479710256801075,0.9995138259991005,0.0413349007990867,0.5206026658066896,0,0.8629755604608232,0.5770411756999720,0.9287420382591846,0.8481530205452407,0.9973319317629893,0.0108969658195211,0.7495690598626537,0.9811385956510490,0.9915991262973569,0.9485404068339514,1,0.7718279184631588])
catools.display_lut(lut)
```

```python
threshold=0.05
```

The configuration below requires $28$ iterations to get all cells equal to 1 and 4 iterations to reach the threshold value.

```python
configuration = catools.parse_2d_configuration("1;1;1;1;0;0;1;0;1", 3, 3)
catools.animate_iterate_configuration_to(configuration, lut, 30, title=catools.title_notify_when_threshold_reached(threshold))
```

The configuration below requires $32$ iterations to get all cells equal to 1 and 8 to reach the threshold.

```python
configuration = catools.parse_2d_configuration("1;1;0;1;1;0;0;1;0", 3, 3)
catools.animate_iterate_configuration_to(configuration, lut, 33, title=catools.title_notify_when_threshold_reached(threshold))
```
